## Module Import & Data Import

```python
#Import listdirectory function from OS module
from os import listdir
import os
import pathlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import cv2
import pandas as pd
import re

from skimage.filters import threshold_otsu
from scipy import ndimage
import math
```

## Batch Import

```python
# #Set directory to home directory which is the 'Image-Processing' folder
p = pathlib.Path('./images/stained_slides/Exp58/2D/')
# Exp58_ESFM_2D_Laminin_10X_1
# Exp58_ESFM_2D_Matrigel_10X_1


# #List subdirectory to get an overview 
data = {'Coating':[], 'Filename': []}
df_images = pd.DataFrame(data) 


#[x for x in p.iterdir()]

filename_list = []
coating_list = []
experiment_list = []
media_list = []
for filename in p.iterdir():
    filename = str(filename).lower()
    #filename_split = split(filename)
    foldername = filename.split("\\")[-1]
    
    if 'exp58_esfm_2d_laminin' in filename:
        #print(filename)
        print(filename.split("\\"))
        print(filename.split("\\")[-1])
        
        media = foldername.split('_')[1]
        coating = foldername.split('_')[3]
        filename_list.append(filename.split("\\")[-1])
        coating_list.append(coating)
        experiment_list.append(filename.split("\\")[2])
        media_list.append(media)
        
    elif 'exp58_esfm_2d_matrigel' in filename:
        media = foldername.split('_')[1]
        coating = foldername.split('_')[3]
        filename_list.append(filename.split("\\")[-1])
        coating_list.append(coating)
        experiment_list.append(filename.split("\\")[2])
        media_list.append(media)
    
    elif 'exp58_esfm_2d_nomatrix' in filename:
        media = foldername.split('_')[1]
        coating = foldername.split('_')[3]
        filename_list.append(filename.split("\\")[-1])
        coating_list.append(coating)
        experiment_list.append(filename.split("\\")[2])
        media_list.append(media)
        
    elif 'exp58_esfm_2D_vitronectin' in filename:
        media = foldername.split('_')[1]
        coating = foldername.split('_')[3]
        filename_list.append(filename.split("\\")[-1])
        coating_list.append(coating)
        experiment_list.append(filename.split("\\")[2])
        media_list.append(media)
        
        #df_images['Filename'] = filename
        
df_images['Filename'] = filename_list
df_images['Coating'] = coating_list
df_images['Experiment'] = experiment_list
df_images['Media'] = media_list
print(df_images)
```

## Plot Original cells

```python
folder_list = df_images['Filename']
list_filenames = []
#print(list_filenames)

fig = plt.figure(figsize = (10,50))
fig.set_facecolor('xkcd:gray')

cols = 2
rows = math.ceil(int(len(folder_list)*2/cols))
counter = 1

image_file_data = {'Filename':[], 'Image_Path':[],'Staining':[]}
image_file_df = pd.DataFrame(image_file_data)

for folder in folder_list:
    for image_path in list(p.glob('**/'+folder+'/*.jpg')):
        image_path = str(image_path) # convert filename into string object to find substring in filename  
          
        if image_path.split("\\")[-1] == 'DAPI.jpg':
            list_filenames.append(('Dapi',image_path,folder))
            img = cv2.imread(str(image_path))  
            ax = fig.add_subplot(rows,cols,counter)
            ax.set_title('Dapi')
            plt.xticks([])
            plt.yticks([])
            plt.imshow(img) 
            counter +=1 
        
        elif image_path.split("\\")[-1] == 'INS.jpg':
            list_filenames.append(('Insulin',image_path,folder))
            img = cv2.imread(str(image_path))  
            ax = fig.add_subplot(rows,cols,counter)
            ax.set_title('Insulin')
            plt.xticks([])
            plt.yticks([])
            plt.imshow(img) 
            counter +=1 
#print('List filenames',list_filenames)  
a, b, c = map(list, zip(*list_filenames))
image_file_df['Filename'] = c
image_file_df['Image_Path'] = b
image_file_df['Staining'] = a

#final_df = pd.concat([image_file_df,df_images], join='outer')
final_df = df_images.merge(image_file_df)

final_df
print(type(final_df))
```

## Convert to and plot grayimage

```python
print(type(final_df))
final_df
```

```python
#Transform to grayimage
grayimages = []
cols = 2
list_imagepaths = list(final_df['Image_Path'])
rows = math.ceil(int(len(list_imagepaths)/cols))
fig = plt.figure(figsize=(10,50))
fig.set_facecolor('xkcd:gray')

for idx, image_path in enumerate(list_imagepaths):
    
    staining = final_df.loc[final_df['Image_Path']==image_path, 'Staining']
    staining = str(staining)
    img = cv2.imread(str(image_path))
    grayimg = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    ax = fig.add_subplot(rows,cols,idx+1)
    ax.set_title('{} (image {})'.format(staining,idx+1))
    plt.xticks([])
    plt.yticks([])
    plt.imshow(grayimg, cmap='gray') 
    
    grayimages.append((staining,grayimg,image_path))    
print(type(final_df))
```

## Plot Greyimage and Mask

```python
maskimages = []

for idx, grayimage in enumerate(grayimages):
    staining_name = grayimage[0]
    grayimage = grayimage[1]
    image_path = grayimage[2]
    thresh_value = threshold_otsu(grayimage)
    maskimg = np.where(grayimage > thresh_value, 1, 0)
    maskimages.append((staining_name, maskimg, image_path))
print(len(maskimages))
```

```python
fig = plt.figure(figsize=(10,50))
fig.set_facecolor('xkcd:gray')
cols = 2
rows = math.ceil(int(len(maskimages)/cols))
count = 1

for idx, maskimage in enumerate(maskimages):
    if count <= len(maskimages):
    #grayimage
        staining_name = grayimages[idx][0]
        ax = fig.add_subplot(rows,cols,count)
        ax.set_title('{} grayimage {}'.format(staining_name, idx+1))
        plt.xticks([])
        plt.yticks([])
        plt.imshow(grayimages[idx][1], cmap='gray')
    
      #mask
        ax = fig.add_subplot(rows,cols,count+1)
        ax.set_title('{} mask {}'.format(staining_name, idx+1))
        plt.xticks([])
        plt.yticks([])
        plt.imshow(maskimage[1], cmap='viridis')
        
        
        count +=2
    else:
        break
```
